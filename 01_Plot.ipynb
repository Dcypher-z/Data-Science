```python
from matplotlib import pyplot as plt

# Dataset 1
import random
x = [i for i in range(1,101)]
y=[];
for i in range(100):
    y.append(random.randint(1,101))
```

```python
#base plot line
plt.plot(x,y)
plt.show()
```

```python
#scatter plot line
plt.scatter(x,y)
plt.show()
```

```python
#labels
plt.plot(x,y)
plt.title('TITLE')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Colors: w r g y k c b m(white, red, green, yellow, black, cyan, blue, magenta)
plt.plot(x,y,'r')
plt.show()
plt.plot(x,y,'k')
plt.show()
```

```python
# Shapes: o p s d h x v . , 1 2 3 4 8 - + | : <> -- -.
plt.plot(x,y,'x')
plt.show()
plt.plot(x,y,'--')
plt.show()
plt.plot(x,y,'8')
plt.show()

#Shapes with colors
plt.plot(x,y,'rx')
plt.show()
plt.plot(x,y,'g--')
plt.show()
```

```python
lines = plt.plot(x,y)
plt.setp(lines, linewidth=3, color='g',linestyle="-.")
plt.show()
```

```python
#plotting multiple datasets on same graph
import random
x = [i for i in range(1,101)]
y_1=[];
y_2=[];
for i in range(100):
    y_1.append(random.randint(1,101))
    y_2.append(random.randint(1,101))
plt.plot(x,y_1,'gx', x, y_2,'ro')
plt.show()
 
plt.plot(x,y_1,'g', x, y_2,'r')
plt.show()
```

```python
# subplots

plt.subplot(2, 1, 1)  # rows, col, index
plt.title('graph 1')
plt.plot(x, y_1, 'r')

plt.subplot(2, 1, 2)
plt.title('graph 2')
plt.plot(x, y_2, 'g')
plt.suptitle("Graphs")
plt.show()
# --------------------------------------
plt.subplot(1, 2, 1)  # rows, col, index
plt.title('graph 1')
plt.plot(x, y_1, 'ro')

plt.subplot(1, 2, 2)
plt.title('graph 2')
plt.plot(x, y_2, 'gx')
plt.show()
# --------------------------------------
```

```python
#Bar graph
bar_chart_data = {'one':5, 'two':4 ,'three':7}
plots = plt.bar(bar_chart_data.keys(), bar_chart_data.values())
plt.setp(plots, color='r') 
plt.title('Bar Graph')
plt.xlabel('category')
plt.ylabel('count')
```

```python
#pie chart
pie_chart_data={'one':2,'two':7,'three':5}
explode=(0.2, 0, 0)
color= ('r', 'w', 'g')
pieplot = plt.pie(pie_chart_data.values(), labels=pie_chart_data.keys(), startangle=0, explode=explode, shadow=True, colors=color)
plt.show()
```

```python
# histogram
histogram_dataset = [3, 4, 3, 5, 4, 7, 1, 4, 2, 2, 7, 4,
                     3, 8, 2, 8, 9, 2, 9, 1, 2, 10, 2, 6, 5, 9, 4, 5, 4, 1]

plt.subplot(2, 1, 1)
plt.hist(histogram_dataset, color='r')
plt.title("Histogram")
plt.xlabel("Number (1-10)")
plt.ylabel("Count")

plt.subplot(2, 1, 2)
plt.hist(histogram_dataset, color='b', cumulative=True)
plt.title("Histogram")
plt.xlabel("Number (1-10)")
plt.ylabel("Count")
plt.show()
```

```python
#3D plot
from mpl_toolkits.mplot3d import Axes3D
axes = plt.axes(projection='3d')
axes.scatter3D(x,y,y_1,marker='x')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_zlabel('z-axis')
plt.show()
```
